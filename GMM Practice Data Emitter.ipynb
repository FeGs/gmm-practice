```python
import math
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import mlab

%matplotlib inline
```

## 남자 키 분포 ~ N(173.25, 6.57^2)

```python
men = np.random.normal(173.25, 6.57, 600)
```

```python
bins = np.linspace(math.ceil(min(men)),
                   math.floor(max(men)),
                   40)

plt.plot(bins, mlab.normpdf(bins, 173.25, 6.57) * 600)
plt.hist(men, bins=bins)

plt.title('20s men height distribution (2010)')
plt.xlabel('height (cm)')
plt.ylabel('count')
```

## 여자 키 분포 ~ N(158.00, 5.05^2)

```python
women = np.random.normal(158.00, 5.05, 200)
```

```python
bins = np.linspace(math.ceil(min(women)),
                   math.floor(max(women)),
                   30)

plt.plot(bins, mlab.normpdf(bins, 158.00, 5.05) * 200)
plt.hist(women, bins=bins)

plt.title('20s men height distribution (2010)')
plt.xlabel('height (cm)')
plt.ylabel('count')
```

```python
observations = np.concatenate([men, women])
```

## 관측값 분포

```python
bins = np.linspace(math.ceil(min(observations)),
                   math.floor(max(observations)),
                   40)

plt.plot(bins, mlab.normpdf(bins, 173.25, 6.57) * 600)
plt.plot(bins, mlab.normpdf(bins, 158.00, 5.05) * 200)
plt.plot(bins, mlab.normpdf(bins, 173.25, 6.57) * 600 + mlab.normpdf(bins, 158.00, 5.05) * 200)
plt.hist(observations, bins=bins)

plt.title('20s height distribution (2010)')
plt.xlabel('height (cm)')
plt.ylabel('count')
```

```python
men_df = pd.DataFrame(men)
```

```python
women_df = pd.DataFrame(women)
```

```python
observations_df = pd.DataFrame(observations)
```

```python
men_df.to_csv('men.csv')
women_df.to_csv('women.csv')
observations_df.to_csv('observations.csv')
```
